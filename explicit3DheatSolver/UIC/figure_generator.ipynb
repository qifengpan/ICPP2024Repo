```python
"""
Put your measurement results here!
"""
FLOPS_DC_tdma = 1.028 # FLOPS of DC on Intel for subroutine kernel of tdma
nSample = [4,8,16,64] #sample size
FLOPS_UIC_tdma = [2.644,4.39,5.064,2.991] # FLOPS of UIC on Intel for subroutine kernel of tdma
AI_UIC_tdma = [0.057,0.059,0.066,0.075] # algorithmic intensity of UIC on Intel for subroutine kernel of compute next

"""
Roofline model drawing
"""

AI_Axis_L2 = np.array([0.001,0.01,0.1,0.26])
Y_L2roof = 226.38*AI_Axis_L2

AI_Axis_L3 = np.array([0.001,0.01,0.1,2.42])
Y_L3roof = 24.22*AI_Axis_L3

X_DRAMroof= np.array([0.001,0.01,0.1,0.11,3.94])
Y_DRAMroof = 14.89*X_DRAMroof


X_ScalarAdd= np.array([0.11,1,10,100,1000,10001,110000])
Y_ScalarAdd= 7.76*np.array([1,1,1,1,1,1,1])

X_VectorAdd= np.array([0.26,1,10,100,1000,10001,110000])
Y_VectorAdd= 58.69*np.array([1,1,1,1,1,1,1])


#L2 bandwidth roofline
plt.plot(AI_Axis_L2,Y_L2roof,"k--")
plt.annotate("L2 Bandwidth",xy=(0.011,3.3),rotation=39,fontsize=8)

#L3 bandwidth roofline
plt.plot(AI_Axis_L3,Y_L3roof,"k--")
plt.annotate("L3 Bandwidth",xy=(0.011,0.35),rotation=39,fontsize=8)
#DRAM bandwidth roofline
plt.plot(X_DRAMroof,Y_DRAMroof,"k--")
plt.annotate("DRAM Bandwidth",xy=(0.011,0.20),rotation=38.5,fontsize=8)
#DP vector Add peak
plt.plot(X_VectorAdd,Y_VectorAdd,"k--")
plt.annotate("DP vector Add peak",xy=(1.5,65),fontsize=8)

##########################################################################
# plt.annotate("Scalar add peak",xy=(1.5,8.2),fontsize=8)
plt.xscale("log")
plt.yscale("log")
markers = ["*" , "," , "o" , "v" ]
Labels  =["UIC: 4 Samples","UIC: 8 Samples","UIC: 16 Samples","UIC: 64 Samples"]




#make a sample dataset
x = AI_UIC_tdma  #test x values.. every feature gets the same x values but you can generalize this
y = FLOPS_UIC_tdma #generate 7 arrays of y values 


for i in range(4): #for each of the 7 features 
    mi = markers[i] #marker for ith feature 
    xi = x[i] #x array for ith feature .. here is where you would generalize      different x for every feature
    yi = y[i] #y array for ith feature 
#     ci = colors[i] #color for ith feature 
    plt.scatter(xi,yi,marker=mi, color="#22A8CA",s = 50,label = Labels[i]) 

plt.scatter(0.104,FLOPS_DC_tdma,marker="^", color= "#A43C1C",s=50,label = "DC") 
plt.xlim([0.01,10])
plt.ylim([0.1,100])
plt.xlabel("FLOP/Byte")
plt.ylabel("GFLOPS")
plt.legend()
plt.show()
```
